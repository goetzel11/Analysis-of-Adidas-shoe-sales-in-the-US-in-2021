# **iv. Data Cleaning**

```python
# Menampilkan Missing Value pada kolom numerik utama sebelum cleaning
print("Missing Value sebelum cleaning:")
print(df[['Price_per_Unit','Unit_Sold','Total_Sales','Operating_Profit']].isnull().sum())
```

```python
# Drop Kolom yang sekiranya tidak relevan dengan data yang ingin dianalisis
df = df.drop(columns=['Retailer_ID', 'City']) # Drop kolom Retailer_ID dan City
```

### **3. Apakah wilayah tertentu memiliki penjualan dan profit yang lebih tinggi dibandingkan wilayah lain?**

```python
# Memastikan kolom Total_Sales dan Operating_Profit dalam format numerik (jaga-jaga)
df['Total_Sales'] = pd.to_numeric(df['Total_Sales'], errors='coerce')
df['Operating_Profit'] = pd.to_numeric(df['Operating_Profit'], errors='coerce')

# menghitung total penjualan dan profit per wilayah
region_summary = df.groupby('Region').agg({ 
    'Total_Sales': 'sum',
    'Operating_Profit': 'sum'
}).sort_values(by='Total_Sales', ascending=False)

# Plot bar penjualan dan profit
fig, ax = plt.subplots(2, 1, figsize=(10, 10))

# Penjualan
region_summary['Total_Sales'].plot(kind='bar', ax=ax[0], title='Total Penjualan per Wilayah', color='skyblue')
ax[0].set_ylabel('Total Penjualan (USD)')
ax[0].ticklabel_format(style='plain', axis='y')

# Profit
region_summary['Operating_Profit'].plot(kind='bar', ax=ax[1], title='Total Profit per Wilayah', color='lightcoral')
ax[1].set_ylabel('Total Profit (USD)')
ax[1].ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.show()

# Menampilkan angka-angkanya
print("Total Penjualan & Profit per Wilayah:")
print(region_summary)
```

### **5. Apakah ada hubungan antara harga per unit dan total unit yang terjual?**

```python

# memastikan kolom Price_per_Unit dan Unit_Sold dalam format numerik (jaga-jaga)
df['Price_per_Unit'] = pd.to_numeric(df['Price_per_Unit'], errors='coerce')
df['Unit_Sold'] = pd.to_numeric(df['Unit_Sold'], errors='coerce')

# menghitung korelasi Spearman antara Price_per_Unit dan Unit_Sold
corr, p_val = spearmanr(df['Price_per_Unit'], df['Unit_Sold']) 

print(f"Korelasi Spearman antara Harga dan Unit Terjual: {corr:.4f}")
print(f"P-value: {p_val:.4e}")

if p_val < 0.05:
    print("Hubungan signifikan secara statistik.")
else:
    print("Hubungan tidak signifikan secara statistik.")

# Interpretasi
if corr < 0:
    print("Artinya: Ada korelasi negatif - harga lebih tinggi cenderung dijual dalam jumlah unit lebih sedikit.")
elif corr > 0:
    print("Artinya: Ada korelasi positif - harga lebih tinggi cenderung dijual dalam jumlah unit lebih banyak.")
else:
    print("Artinya: Tidak ada korelasi.")
```

Saya mengunakan statistika inferensial korelasi Spearman karena ingin menguji apakah ada hubungan antara harga dan jumlah unit terjual, dan apakah hubungan itu signifikan secara statistik. Spearman dipilih karena lebih fleksibel dan tidak terlalu sensitif terhadap data yang tidak linear.
kemudian juga menggunakan hipotesa untuk uji korelasi Spearman karena ingin menguji apakah ada hubungan antara harga dan jumlah unit terjual, dan apakah hubungan itu signifikan secara statistik. Saya menentukan hipotesis nol (H0) bahwa tidak ada hubungan, dan hipotesis alternatif (H1) bahwa ada hubungan. Dengan p-value < 0.05, saya menolak H0 dan menyimpulkan bahwa ada hubungan yang signifikan.

#### **Insigth**

Dari uji korelasi Spearman, kita temukan bahwa ada hubungan positif yang signifikan secara statistik antara harga dan jumlah unit terjual. Ini artinya, produk dengan harga lebih tinggi justru cenderung terjual dalam jumlah unit yang lebih banyak. Ini pola yang tidak biasa, dan bisa disebabkan oleh produk premium atau strategi pricing tertentu.
